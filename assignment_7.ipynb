## 1. Dice and Roulette

**(a) Expected value of a single die roll**

For a fair six-sided die,
\[
\mathbb{E}[X] = \frac{1+2+3+4+5+6}{6} = 3.5.
\]

**(b) Expected value of rolling two dice**

Let \(X_1\) and \(X_2\) be independent. Then
\[
\mathbb{E}[X_1 + X_2] = 3.5 + 3.5 = 7.
\]

**(c) Expected winnings in European roulette**

A \$1 bet wins with probability \(\frac{K}{37}\) and pays \(\frac{36}{K}\).  
\[
\mathbb{E}[\text{payoff}] = \frac{K}{37} \cdot \frac{36}{K} = \frac{36}{37}.
\]
Net expectation:
\[
\mathbb{E}[W] = \frac{36}{37} - 1 = -\frac{1}{37} \approx -0.027.
\]

**(d) Roll again if the first roll is a six**

\[
T = X_1 + \mathbf{1}\{X_1 = 6\} X_2.
\]
\[
\mathbb{E}[T] = 3.5 + \frac{1}{6}\cdot 3.5 = \frac{24.5}{6} \approx 4.0833.
\]

**(e) Repeat this process until you fail to roll a six**

Let \(J\) be the expected total.
\[
J = \sum_{k=1}^5 k\cdot \frac{1}{6} + \frac{1}{6}(6+J)
= \frac{15}{6} + 1 + \frac{J}{6}
= \frac{21}{6} + \frac{J}{6}.
\]
Solving:
\[
J = \frac{21}{5} = 4.2.
\]

## 4. Covariance and Independence

\[
\mathrm{cov}(X,Y) = \mathbb{E}[(X-\mu_X)(Y-\mu_Y)].
\]

**(a) Independence implies zero covariance**

If \(f_{XY}(x,y)=f_X(x)f_Y(y)\),
\[
\mathrm{cov}(X,Y)
= \left(\int (x-\mu_X)f_X(x)\,dx\right)
  \left(\int (y-\mu_Y)f_Y(y)\,dy\right)
= 0.
\]

**(b) Zero covariance does not imply independence**

Let \(X\sim\text{Uniform}(-1,1)\), \(Y=X^2\).

\[
\mathbb{E}[X]=0, \quad \mathbb{E}[X^3]=0.
\]

Thus
\[
\mathbb{E}[XY] = \mathbb{E}[X^3] = 0
\Rightarrow \mathrm{cov}(X,Y)=0.
\]

But \(Y\) is completely determined by \(X\), so they are not independent.

## 5. Centering and Standardizing Random Variables

Assume \(\mathbb{E}[X]=\mu\) and \(\mathrm{Var}(X)=\sigma_X^2\).

**(a)** Let \(\varepsilon = X - \mu\). Then
\[
\mathbb{E}[\varepsilon] = 0.
\]

**(b)**  
\[
\mathrm{Var}(\varepsilon) = \mathbb{E}[(X-\mu)^2] = \sigma_X^2.
\]

So
\[
X = \mu + \varepsilon.
\]

**(c)** Define the standardized variable
\[
Z = \frac{X-\mu}{\sigma_X}.
\]

Then
\[
\mathbb{E}[Z]=0,\quad \mathrm{Var}(Z)=1.
\]

Thus
\[
X = \mu + \sigma_X Z.
\]

This expresses any finite-variance random variable as a mean plus a scaled standardized noise term.


## 5. Centering and Standardizing Random Variables

Assume
\[
\mathbb{E}[X] = \mu, \qquad \mathrm{Var}(X) = \sigma_X^2 < \infty.
\]

### (a) Define \(\varepsilon = X - \mathbb{E}[X]\). Find \(\mathbb{E}[\varepsilon]\)

By construction,
\[
\mathbb{E}[\varepsilon]
= \mathbb{E}[X - \mu]
= \mathbb{E}[X] - \mu
= \mu - \mu
= 0.
\]

So \(\varepsilon\) is a mean-zero version of \(X\).

---

### (b) Find \(\mathrm{Var}(\varepsilon)\)

\[
\mathrm{Var}(\varepsilon)
= \mathbb{E}[(\varepsilon - \mathbb{E}[\varepsilon])^2]
= \mathbb{E}[(X - \mu - 0)^2]
= \mathbb{E}[(X - \mu)^2]
= \mathrm{Var}(X)
= \sigma_X^2.
\]

Thus we can write
\[
X = \mu + \varepsilon,
\quad
\mathbb{E}[\varepsilon] = 0,
\quad
\mathrm{Var}(\varepsilon) = \sigma_X^2.
\]

---

### (c) Show we can write \(X = \mu + \sigma_X Z\) with \(\mathbb{E}[Z]=0\) and \(\mathrm{Var}(Z)=1\)

Define the standardized variable
\[
Z = \frac{X - \mu}{\sigma_X} = \frac{\varepsilon}{\sigma_X}.
\]

Then
\[
\mathbb{E}[Z]
= \frac{1}{\sigma_X}\mathbb{E}[\varepsilon]
= \frac{1}{\sigma_X} \cdot 0
= 0,
\]
and
\[
\mathrm{Var}(Z)
= \mathrm{Var}\left(\frac{\varepsilon}{\sigma_X}\right)
= \frac{1}{\sigma_X^2}\mathrm{Var}(\varepsilon)
= \frac{1}{\sigma_X^2} \cdot \sigma_X^2
= 1.
\]

Rewriting,
\[
X = \mu + \sigma_X Z,
\]
where \(Z\) has mean 0 and variance 1.

This expresses any finite-variance random variable as a mean plus a scaled standardized “noise” term, which is the same structure used in many regression and modeling setups.
